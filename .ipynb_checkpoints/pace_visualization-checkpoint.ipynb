Only up to 12 chunks here with. Best of 5 results.

````python
data_list = [
    pandas.read_csv('slurm-299016.csv', index_col=False),
    pandas.read_csv('job1/slurm-299018.csv', index_col=False)
]

data = pandas.concat(data_list);
````

## Full Pipeline

````python
read = lambda x: pandas.read_csv(x, index_col=False)

data = pandas.concat(fmap(read, ['job5/slurm-299131.csv', 'job6/slurm-299143.csv']))

data
````

````python
@curry
def munge(data_, n_samples):
    df = data_.loc[data.samples == n_samples]
    fit_control = df.query('chunks == 1').fit_time.median()
    predict_control = df.query('chunks == 1').predict_time.median()
    new_df = df.assign(fit_speed_up=fit_control / df.fit_time, predict_speed_up=predict_control / df.predict_time)
    return new_df.groupby('chunks').agg(dict(
        samples='min',
        fit_time='median',
        predict_time='median',
        fit_speed_up=['median', 'std'],
        predict_speed_up=['median', 'std']
    ))
    
````

````python
data_list = list(fmap(munge(data), data.samples.unique()))

data_list[0]
````

````python
plt.style.use('ggplot')

plt.figure(figsize=(10, 8))

def plot_fit(df):
    x = df.index
    y = df.fit_speed_up['median']
    e = df.fit_speed_up['std']
    plt.loglog(x, y, label=f'{df.samples["min"].iloc[0]} Samples')
    plt.fill_between(x, y - e, y + e, alpha=0.1)

for data in data_list:
    plot_fit(data)

plt.loglog([1, 48], [1, 48], lw=2, label='Perfect')
plt.title('Chunks v Speed Up (fit)')
plt.xlabel('Chunks')
plt.ylabel('Speed Up')
plt.legend()
    
````

````python
plt.style.use('ggplot')


plt.figure(figsize=(10, 8))

def plot_fit(df):
    x = df.index
    y = df.predict_speed_up['median']
    e = df.predict_speed_up['std']
    plt.loglog(x, y, label=f'{df.samples["min"].iloc[0]} Samples')
    plt.fill_between(x, y - e, y + e, alpha=0.1)

for data in data_list:
    plot_fit(data)

plt.loglog([1, 48], [1, 48], lw=2, label='Perfect')
plt.title('Chunks v Speed Up (predict)')
plt.xlabel('Chunks')
plt.ylabel('Speed Up')
plt.legend()
````

## Typical submit file

```
#!/bin/bash
#SBATCH --partition=rack3               # -p, first available from the list of partitions
#SBATCH --time=04:00:00                 # -t, time (hh:mm:ss or dd-hh:mm:ss)
#SBATCH --nodes=1                       # -N, total number of machines
#SBATCH --ntasks=48                      # -n, 64 MPI ranks per Opteron machine
#SBATCH --cpus-per-task=1               # threads per MPI rank
#SBATCH --job-name=pca                  # -J, for your records
#SBATCH --workdir=/working/wd15/git/PyMKS-Efficiency-Paper/slurm/job5 # -D, full path to an existing directory
#SBATCH --qos=test
#SBATCH --mem=0G

filename=slurm-${SLURM_JOB_ID}.csv

echo "chunks,samples,fit_time,predict_time" > ${filename}

for n_chunk in 1 2 4 8 16 24 32 40 48
do
    for repeat in {1..5}
    do
        for samples in 1000 2000 4000
        do		       
            python pca.py --samples=${samples} --chunks=${n_chunk} --filename=${filename} --threads_per_worker=1 --n_workers=48
	done
    done
done	
```

````python
read = lambda x: pandas.read_csv(x, index_col=False)

df = pandas.concat(fmap(read, ['job5/slurm-299131.csv', 'job6/slurm-299143.csv']))

df[:10]
````

````python
df_ = df.groupby(['chunks', 'samples']).agg(dict(
    fit_time=['median', 'std'],
    predict_time=['median', 'std']
))

df_
````

````python
df_.index.get_level_values(0).unique()
````

````python
plt.style.use('ggplot')

plt.figure(figsize=(10, 8))

for ind in df_.index.get_level_values(0).unique():
    plt.plot(df_.loc[ind].index, df_.loc[ind].fit_time['median'], label=f"{ind} chunks")
    
plt.title('Samples v Fit Time (48 workers)')
plt.xlabel('Samples')
plt.ylabel('Fit Time (s)')
plt.legend()
````
